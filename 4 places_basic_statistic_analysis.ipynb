```python
sagrada_clean_dataset = pd.read_csv(r'clean_dataset\sagrada_clean_dataset.csv', encoding='utf-8',index_col=0)
mila_clean_dataset = pd.read_csv(r'clean_dataset\mila_clean_dataset.csv', encoding='utf-8',index_col=0)
parkguell_clean_dataset = pd.read_csv(r'clean_dataset\parkguell_clean_dataset.csv', encoding='utf-8',index_col=0)
batllo_clean_dataset = pd.read_csv(r'clean_dataset\batllo_clean_dataset.csv', encoding='utf-8',index_col=0)
```

```python
nums = [len(parkguell_clean_dataset),len(batllo_clean_dataset),len(mila_clean_dataset),len(sagrada_clean_dataset)]

df = pd.DataFrame(nums, index =['Parque Güell','Casa Batllò', 
                                'Casa Milà', 
                                'Sagrada Família'])
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
plt.style.use('seaborn-paper')
ax = df.plot(kind='barh', figsize=(8.5,4.5))
ax.get_legend().remove()
ax.set_title('Cantidad de Posts mencionados a cada patrimonios culturales \n en el red social Weibo (2017 - 2020) \n')

for i, v in enumerate(nums):
    ax.text(x=v-1200, y=i, s=str(v),verticalalignment="center", color="w",horizontalalignment="left", size=9)
    
ax.set_xlabel('Número de Post')
ax.plot()
plt.show()
```
